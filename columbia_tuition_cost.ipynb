```python
import pandas as pd
```

```python
df = pd.read_csv('columbia_ms_tuition2.csv')
df.info()
```

```python
df.describe()
```

```python
df["tuition_per_year"] = df["tuition_per_year"].str.replace("$", "", regex=False).str.replace(",", "").astype(float)
df["estimated_total_cost"] = df["estimated_total_cost"].str.replace("$", "", regex=False).str.replace(",", "").astype(float)
```

```python
df.describe()
```

```python
df.sort_values(by='tuition_per_year', ascending=False)
```

```python
df.sort_values(by='estimated_total_cost', ascending=False)
```
