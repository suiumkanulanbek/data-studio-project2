**Information about the dataset**

This dataset was taken from the **US Bureau of Labor Statistics**. 
The dataset is called **"Median weekly earnings of full-time wage and salary workers by detailed occupation and sex"**.
Numbers are shown in thousands. Link to the original dataset: https://www.bls.gov/cps/cpsaat39.htm

```python
import pandas as pd
```

```python
df = pd.read_csv('median_weekly_wage.csv')
df.info()
```

```python
df.describe()
```

```python
df.sort_values(by='mdn_weekly_earnings_total', ascending=False).head(10)
```

```python
columns_to_clean = ["mdn_weekly_earnings_total", "mdn_weekly_earnings_men", "mdn_weekly_earning_women"]

dash_variants = ["-", "–", "—"]  

df[columns_to_clean] = (
    df[columns_to_clean]
    .replace({'\$': '', ',': '', '|'.join(dash_variants): '0'}, regex=True) 
    .astype(float)  
)
```

```python
df.sort_values(by='mdn_weekly_earnings_total', ascending=False).head(10)
```
